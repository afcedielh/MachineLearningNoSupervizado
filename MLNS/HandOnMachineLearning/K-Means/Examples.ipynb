## Celda 1: Importación de Módulos Esenciales para K-Means

**Propósito General:**
Esta celda importa los módulos necesarios del ecosistema scikit-learn para implementar el algoritmo de agrupamiento K-Means y generar conjuntos de datos sintéticos de prueba.

**Explicación Línea por Línea:**

- `from sklearn.cluster import KMeans`
   Importa la clase `KMeans` del módulo `sklearn.cluster`. Esta clase implementa el algoritmo K-Means, un método de aprendizaje no supervisado que particiona un conjunto de datos en `k` clústeres mediante la minimización iterativa de la varianza intraclúster (suma de cuadrados intragrupo). `KMeans` es fundamental para tareas de segmentación de datos, reducción de dimensionalidad y análisis exploratorio.

- `from sklearn.datasets import make_blobs`
   Importa la función `make_blobs` del módulo `sklearn.datasets`. Esta función genera conjuntos de datos sintéticos compuestos por puntos (blobs) distribuidos aleatoriamente alrededor de centros especificados. Es extremadamente útil para pruebas, validación y demostración de algoritmos de clustering, ya que permite controlar la estructura conocida del dataset.


```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

 ## Celda 2: Generación del Conjunto de Datos Sintético

 **Propósito General:**
 Esta celda genera un conjunto de datos sintético mediante la función `make_blobs`, que produce puntos distribuidos en clústeres gaussianos. Este dataset se utilizará como entrada para entrenar el algoritmo K-Means.

 **Explicación Línea por Línea:**

- `x, y = make_blobs(n_samples=300, centers=5, cluster_std=0.60, random_state=42)`
   Invoca la función `make_blobs` con parámetros específicos y retorna dos arrays:
   - `x`: matriz de características con forma `(300, 2)` que contiene las coordenadas de los puntos en el espacio euclidiano bidimensional.
   - `y`: vector de etiquetas verdaderas (ground truth) con forma `(300,)` que asigna cada punto a su clúster real (valores 0 a 4).

 - `#x, y = make_blobs([...])`
   Línea comentada que sugiere código alternativo. Se mantiene como referencia de flexibilidad en la generación de datos.

 **Análisis de Parámetros e Hiperparámetros:**

 | Parámetro | Valor | Definición y Impacto |
 |-----------|-------|---------------------|
 | `n_samples` | `300` | Número total de puntos de datos a generar. Controla el tamaño del dataset. Un valor de 300 proporciona una cantidad suficiente de puntos para entrenar el modelo sin ser computacionalmente intensivo. Mayor cantidad mejora la estabilidad estadística pero incrementa el tiempo de cómputo. |
 | `centers` | `5` | Número de centros gaussianos alrededor de los cuales se distribuyen los puntos. Cada centro define una distribución normal multivariada independiente. Con 5 centros, se espera que el algoritmo K-Means descubra 5 clústeres naturales en los datos. |
 | `cluster_std` | `0.60` | Desviación estándar de cada clúster gaussiano. Controla la dispersión o compactación de los puntos alrededor de cada centro. Un valor de 0.60 indica una varianza moderada: los clústeres son distinguibles pero con cierta sobreposición. Valores menores (ej. 0.1) producen clústeres más compactos; valores mayores (ej. 2.0) producen clústeres más dispersos y superpuestos. |
 | `random_state` | `42` | Semilla del generador pseudoaleatorio (seed). Fija el valor a 42 asegura reproducibilidad absoluta: cada ejecución del código generará exactamente el mismo dataset. Crítico para validación y debugging. Sin este parámetro, los datos variarían en cada ejecución. |


```python
x, y = make_blobs(n_samples=300, centers=5, cluster_std=0.60, random_state=42)
```

 ## Celda 3: Inicialización, Entrenamiento y Predicción con K-Means

 **Propósito General:**
 Esta celda instancia el modelo K-Means, lo entrena sobre los datos y genera predicciones de clústeres para cada punto del dataset.

 **Explicación Línea por Línea:**

 - `k = 5`
   Asigna la variable local `k` con el valor 5. Esta variable define el número de clústeres que se utilizará en el algoritmo K-Means, coincidiendo intencionalmente con el número de `centers` utilizado en la generación de datos. Aunque K-Means no requiere conocimiento previo del número real de clústeres, en este caso se ha seleccionado el valor óptimo para evaluar la capacidad del algoritmo.

 - `kmeans = KMeans(n_clusters=k, random_state=42)`
   Crea una instancia del estimador `KMeans` con configuración específica. Este objeto encapsula toda la lógica del algoritmo K-Means (inicialización, asignación iterativa de puntos a centros y actualización de centroides). No realiza ningún cálculo en esta línea; solo configura el modelo.

 - `y_pred = kmeans.fit_predict(x)`
   Método que combina dos operaciones en una sola llamada:
   1. `fit(x)`: entrena el modelo K-Means sobre los datos `x`, inicializando centroides, asignando puntos iterativamente a clústeres y optimizando la posición de los centros hasta convergencia.
   2. `predict(x)`: asigna cada punto en `x` a su clúster más cercano (usando distancia euclidiana).
   El resultado es un array `y_pred` de forma `(300,)` que contiene etiquetas predichas (0 a 4) para cada punto.

 - `y_pred`
   Línea que muestra (output) el contenido del array `y_pred`. En un Jupyter Notebook, esta línea sin asignación imprime el resultado de la evaluación de la variable.

 **Análisis de Parámetros e Hiperparámetros:**

 | Parámetro | Valor | Definición y Impacto |
 |-----------|-------|---------------------|
 | `n_clusters` | `5` | Número de clústeres a descubrir. Es el hiperparámetro crítico de K-Means. Un valor de 5 debe coincidir con la estructura verdadera del dataset para lograr convergencia óptima. Si `n_clusters` fuera menor (ej. 3), el algoritmo fusionaría clústeres reales, incrementando el error. Si fuera mayor (ej. 7), fragmentaría clústeres reales en subestructuras artificiales. |
 | `random_state` | `42` | Semilla para la inicialización de centroides. K-Means es sensible a la inicialización inicial: diferentes posiciones iniciales de centroides pueden conducir a mínimos locales distintos. Fijar `random_state=42` asegura que los centroides se inicialicen siempre en las mismas posiciones pseudoaleatorias, garantizando reproducibilidad. |


```python
k = 5
kmeans = KMeans(n_clusters=k, random_state=42)
y_pred = kmeans.fit_predict(x)
y_pred
```

 ## Celda 4: Verificación de Identidad entre Predicciones y Atributo `labels_`

 **Propósito General:**
 Esta celda verifica que el vector de etiquetas predichas retornado por `fit_predict()` es idéntico al atributo `labels_` almacenado internamente en el objeto `kmeans` tras el entrenamiento. Esto valida la consistencia del modelo.

 **Explicación Línea por Línea:**

 - `y_pred is kmeans.labels_`
   Utiliza el operador `is` de Python para verificar si `y_pred` y `kmeans.labels_` hacen referencia al mismo objeto en memoria (identidad de objetos). En scikit-learn, `fit_predict()` retorna el atributo `labels_` que se almacena internamente tras el entrenamiento. Por lo tanto, esta expresión debería evaluar a `True`, confirmando que no se realiza copia de datos, sino que se retorna una referencia al mismo array. Este comportamiento es importante para eficiencia de memoria y coherencia interna.

 **Análisis Técnico:**

 - **Operador `is` vs. `==`**: El operador `is` verifica identidad (si dos variables apuntan a la misma ubicación de memoria), mientras que `==` verifica igualdad (si los valores contenidos son los mismos). En este caso, `is` es más restrictivo y apropiado para verificar si scikit-learn internamente retorna la misma referencia sin duplicación.

 - **Implicaciones**: Si el resultado es `True`, confirma que `kmeans.labels_` es el único atributo de referencia de las predicciones. Si fuera `False`, indicaría que se realizó una copia, lo cual sería ineficiente. En la práctica, scikit-learn mantiene este comportamiento por optimización.


```python
y_pred is kmeans.labels_
```

 ## Celda 5: Extracción e Impresión de Centros de Clústeres

 **Propósito General:**
 Esta celda accede y visualiza los centros (centroides) finales de los clústeres que el algoritmo K-Means ha optimizado tras la convergencia durante el entrenamiento. Estos centros representan la localización espacial representativa de cada clúster en el espacio de características.

 **Explicación Línea por Línea:**

 - `print("Centros de los clústeres:")`
   Imprime una cadena de texto en la consola como encabezado descriptivo, indicando que los datos a continuación corresponden a las coordenadas de los centros de clústeres.

 - `print(kmeans.cluster_centers_)`
   Accede al atributo `cluster_centers_` del objeto `kmeans`, que es un array NumPy de forma `(5, 2)`. Cada fila representa un centroide (centro geométrico) de un clúster:
   - Dimensión 1: 5 centros (correspondientes a `n_clusters=5`).
   - Dimensión 2: 2 características (dimensionalidad del espacio de características original).
   Cada fila contiene las coordenadas cartesianas `(x, y)` del centroide en cuestión. Estos valores se optimizaron iterativamente mediante el algoritmo K-Means para minimizar la varianza intraclúster (suma de distancias euclidianas al cuadrado entre cada punto y su centroide asignado).

 **Análisis Técnico:**

 - **Atributo `cluster_centers_`**: Este atributo solo está disponible después de invocar `fit()` o `fit_predict()`. Contiene las posiciones finales de los centroides tras la convergencia del algoritmo. En K-Means, la convergencia ocurre cuando las asignaciones de puntos a clústeres ya no cambian, o se alcanza el número máximo de iteraciones especificado.

 - **Interpretación de valores**: Si el dataset fue generado con `centers=5` y `cluster_std=0.60`, se espera que `cluster_centers_` contenga valores próximos (pero no idénticos) a los centros verdaderos del dataset original, ya que el algoritmo K-Means las ha descubierto empíricamente a partir de los datos.

 - **Uso práctico**: Los centros de clústeres son fundamentales para:
   - Interpretación de clústeres (análisis de qué características definen cada grupo).
   - Asignación de nuevos datos: cualquier punto nuevo se asigna al clúster cuyo centroide está más próximo (usando distancia euclidiana).
   - Visualización: en espacios 2D o 3D, los centros actúan como representantes visuales de los clústeres.


```python
print("Centros de los clústeres:")
print(kmeans.cluster_centers_)
```
